```python
import pandas as pd
import os

# Load Excel file
excel_file = "bilingual.xlsx"
df = pd.read_excel(excel_file)

# Column you want to extract
column_name = "transcript"

# Output directory
output_dir = "./"
os.makedirs(output_dir, exist_ok=True)

# Create one txt file per row
for index, value in enumerate(df[column_name]):
    if pd.notna(value):  # skip empty cells
        file_path = os.path.join(output_dir, f"audio_{index+1}.txt")
        
        with open(file_path, "w", encoding="utf-8") as f:
            f.write(str(value))

print("Text files created successfully from Excel!")
```
